```python
# creating histogram for EDA 

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
temp = pd.read_csv('temperature.csv')
```

```python
temp
x=temp.iloc[:,[0]].values
```

```python
plt.title("Distribution of temperature") 
plt.xlabel("Temperature/degree C") 
plt.ylabel("Frequency")
plt.hist(x,bins=20)
```

```python
# creating visualization for scatter plot 
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
plt.scatter(x,y)
plt.show()
```

```python
# boxplot 
import matplotlib.pyplot as plt
data = [5,16,21,26,30,31,32,35,38,66]
data2 = [5,18,25,28,30,33,40,45,50,66]
plt.boxplot([data,data2])
plt.title('Boxplot')
plt.xlabel('TV viewing times (hrs)')
plt.show()
```

```python
# calculate the correlation matrix
corr = df.corr()
print(corr)
```

```python
# plot the heatmap
import seaborn as sns
sns.heatmap(corr, 
xticklabels=corr.columns.values,
yticklabels=corr.columns.values)
plt.show()
```
