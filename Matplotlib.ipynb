```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y)
```

```python
plt.plot(x,y, label = 'linear')
plt.legend(loc=0)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,1,4,6,2,7,8,4,8]
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,label='group 1')
plt.legend(loc=0)
plt.title("this is a scatterplot") 
plt.xlabel("x-label") 
plt.ylabel("y-label")
plt.show()
```

```python
data=np.random.normal(3,1,1000)
plt.hist(data)
```
